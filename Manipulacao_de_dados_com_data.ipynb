#Qual categoria de produto tem o maior ticket médio?

```python
import pandas as pd
import os
```

```python
os.listdir('./')
```

```python
produto_categoria = pd.read_csv(f'/content/olist_products_dataset.csv',
                                usecols=['product_id', 'product_category_name'],
                                index_col='product_id')
produto_categoria
```

```python
pedido_produto = pd.read_csv(f'/content/olist_order_items_dataset.csv', usecols=['order_id','product_id', 'price', 'order_item_id'], index_col='product_id')
pedido_produto
```

```python
#join
pedido_produto.join(produto_categoria, how='inner')
```

```python
#merge
categoria_preco_do_produto_por_pedido = pd.merge(pedido_produto, produto_categoria, on='product_id', how='inner')
categoria_preco_do_produto_por_pedido
```

```python
# respondendo o ticket media das categorias do produto
categoria_preco_do_produto_por_pedido.groupby('product_category_name')['price'].mean().sort_values()
```

```python
#pegando o top 10
categoria_preco_do_produto_por_pedido.groupby('product_category_name')['price'].mean().sort_values(ascending=False).head(10)
```
